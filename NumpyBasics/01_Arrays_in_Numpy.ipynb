# 1. Numpy Arrays

!!! Please note all the elements of Numpy array must be of the same type !!!


```python
import numpy as np

# Array
arr1 = np.array([1, 2, 3, 4, 5])
print("\nArray:", arr1)
print("Elements type:", arr1.dtype)

# Array of Zeros
arr2 = np.zeros(5)
print("\nArray of zeros:", arr2)
print("Elements type:", arr2.dtype)

# Array of Ones
arr3 = np.ones(5)
print("\nArray of ones:", arr3)
print("Elements type:", arr3.dtype)

# Array of Random Numbers
arr4 = np.random.random(5)
print("\nArray of randoms:", arr4)
print("Elements type:", arr4.dtype)

# Array of Normal Random Variables
arr5 = np.random.randn(5)
print("\nArray of standard normal variables:", arr5)
print("Elements type:", arr5.dtype)

# Array of Equally Spaced Elements
arr6 = np.linspace(0, 10, 5)
print("\nArray of equally distributed values:", arr6)
print("Elements type:", arr6.dtype)

# Array by Range
arr7 = np.arange(0, 5, 0.5)
print("\nArray of elements in range:", arr7)
print("Elements type:", arr7.dtype)
```

### Selecting Multiple Elements by Index

```python
import numpy as np

# Array
arr = np.array([10, 12, 1, 7, 22, 14, 11])
print("Array:", arr)

# Accessing Elements by Index
print("arr[0] =", arr[0])
print("arr[3] =", arr[3])

# Select elemens with indices 0, 2, 4, and 6
arr[[0, 2, 4, 6]]
```

```python
import numpy as np

# Array
arr = np.array([10, 12, 1, 7, 22, 14, 11])
print("Array:", arr)

# Array Filtering Principle
arr[[True, False, False, True, False, False, True]]

# Array Filtering Example: Filter elements less than or equal to 10
arr1 = arr[arr <= 10]
print(arr1)
```

## 2.3 Dependend Data Versus Independent Data

### 2.3.1 Generating Independent Data

```python
# Number of data points
N = 100

# Vector X
vx2 = np.linspace(-10, 10, N)

# Vector Y does not depend on X
vy2 = np.random.randn(100)
```

### 2.3.2 Generating Linearly Dependent Data

```python
# Number of points
N = 100

# Vector X
vx3 = np.linspace(-10, 10, N)

# Vector Y
vy3 = [2*x + 8 + 3*np.random.randn(1) for x in vx2]

# Fitting Linear Regression vy = f(vx)
a, b = np.polyfit(vx3, vy3, 1)

vfit = a * vx3 + b
```

### 2.3.3 Plotting Independent Data

```python
import matplotlib.pyplot as plt

# Plotting Independen Variables X and Y

plt.figure(figsize=(10, 6))
plt.scatter(vx2, vy2, s=30, linewidth=3, color="steelblue", alpha=0.5)

# Plot title
plt.title("\nIndependent variables X and Y\n", fontsize=16, weight="bold")

# Axis Labels
plt.xlabel("X", fontsize=16, weight="bold")
plt.ylabel("Y", fontsize=16, weight="bold")


# X ir Y axis
plt.axvline(x=0, linestyle="solid", color="orangered", linewidth=1)  
plt.axhline(y=0, linestyle="solid", color="orangered", linewidth=1)

# Grid
plt.grid()

# Axis font adjustments
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Display plot
plt.show()
```

### 2.3.4 Plotting Dependent Data

```python
import matplotlib.pyplot as plt

# Plot
plt.figure(figsize=(14, 8))
plt.scatter(vx3, vy3, s=50, color="orangered")

# Plot title
plt.title("\nLinear Dependency\n", fontsize=18, weight="bold")

# X-axis and Y-axis labels
plt.xlabel("X", fontsize=16, weight="bold")
plt.ylabel("Y", fontsize=16, weight="bold")


# Adding X and Y axis to the plot
plt.axvline(x=0, linestyle="solid", color="orangered", linewidth=1)  
plt.axhline(y=0, linestyle="solid", color="orangered", linewidth=1)

# Grid
plt.grid()

# Axis font size adjustments
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Adding fitted regression line to the plot
plt.plot(vx3, vfit, color="steelblue", linewidth=3, alpha=0.35)

plt.show()
```

# 3. Histogram Plots

## 3.1 What distribution np.random.random() relies upon?

```python
import numpy as np

random_points = np.random.random(1000)
plt.hist(random_points, bins=10, density=False, edgecolor="white")
plt.title("\nHistogram generated from np.random.random()\n")
plt.show()

# Sample Mean
sample_mean = np.mean(random_points)
print("Sample Mean:", sample_mean)

# Sample Standard Deviation
sample_std = np.std(random_points)
print("Sample Standar Deviation:", sample_std)

# Sample Variance
sample_variance = np.var(random_points)
print("Sample Variance:", sample_variance)
```

**Answer:** np.random.random() takes values from uniform distribution on [0, 1]

## 3.2. Standard Normal Distribution

```python
import numpy as np

# Statndard Normal Distribution Function

def normal_df(x, mu=0, sigma=1): 
    return 1 / sigma / np.sqrt(2*np.pi) * np.exp(-1/2 * ((x - mu)/sigma)**2)

vectorized_normal_df = np.vectorize(normal_df)

x = np.linspace(-3.5, 3.5, 100)
y = vectorized_normal_df(x)


# Sample from Standard Norma Distribution
stnorm_points = np.random.normal(loc=0, scale=1, size=1000)
plt.hist(stnorm_points, bins=15, density=True, label="sample histogram")
plt.title("\nStandard Normal Distribution Function \nvs\n Sample Histogram\n")

# Add Standard Normal Distribution Density Function
plt.plot(x, y, label="standard normal density")

# Add grid
plt.grid()

plt.legend()

# Display Plot
plt.show() 
```
