Visų pirma turime importuoti **matplotlib** biblioteką. Taip pat importuosime NumPy biblioteką ir nauduosime NumPy funkcijas duomenų generavimui.

```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2 style="color:seagreen">How to plot a function?</h2>

1. Create a function $f(x) = 17 \times x^3 - 25 \times x^2 + 28 \times x - 100$ 
2. Create a list x = [-10, -5, -2, 0, 2, 5, 10]
3. Calculate the value $f(x)$ at every point of $x$ 
4. Create a plot of $f(x)$

```python
# Let's define a function

def f(x):
    return 17 * x**3 - 25 * x**2 + 28 * x - 100
```

```python
# Let's define points on x axis (we will calculate f(x) for each of these points)

vx = [-10, -5, -2, 0, 2, 5, 10]
```

```python
# Let's calculat f(x) at every point of vx

vy = [] 

for point in vx:
    # calculate y(x)
    y = f(point)
    # save calculation result to an array
    vy.append(y)
    
print(vy)
```

```python
# Let's make a plot of f(x)

plt.figure(figsize=(14, 8))

plt.plot(vx, vy, linewidth=3, color="steelblue")
plt.title("\nFunction's Plot: y = f(x)\n", fontsize=22, weight="bold")
plt.xlabel("X", fontsize=16, weight="bold")
plt.ylabel("Y", fontsize=16, weight="bold")


# X ir Y ašys
plt.axvline(x=0, linestyle="solid", color="red", linewidth=1)  
plt.axhline(y=0, linestyle="solid", color="red", linewidth=1)

# Tinklelis
plt.grid()

# Ašių skalės
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Pavaizduokime X vektoriaus taškus
plt.scatter(vx, vy, s=100, color="lime")

plt.show()
```

**Problem:** the function $f(x) = 17 \times x^3 - 25 \times x^2 + 28 \times x - 100$ must be smooth and continuous, and our graph doesn't look like that at all. :(
yra tolydi ir nepertraukiama kreivė, o mūsų grafike yra pavaizduota laužtinė. :( 

**Question:** Can we improve our graph? How?

**Answer:** Yes, we can! We just need to use more points when plotting. The more points we are using, the more preces graph we are getting.

<h2 style="color:green">Improving the Plot</h2>

Let's devide the interval [-10, 10] into 100 equal parts, and let's redraw our graph.


```python
# Vector X

vx1 = np.linspace(-10, 10, 101)

# Vector Y

vy1 = [] 

# Let's calculate the values of f(x) at every point of vx1 and save the result to the vy1
for point in vx1:
    ypoint = f(point)
    vy1.append(ypoint)

    
```

```python
# Plotting the dependency: y = f(x) 

plt.figure(figsize=(14, 8))

plt.plot(vx1, vy1, linewidth=3, color="steelblue")
plt.title("\nFunction: Y = f(X)\n", fontsize=22, weight="bold")
plt.xlabel("X", fontsize=16, weight="bold")
plt.ylabel("Y", fontsize=16, weight="bold")


# X ir Y ašys
plt.axvline(x=0, linestyle="solid", color="orangered", linewidth=1)  
plt.axhline(y=0, linestyle="solid", color="orangered", linewidth=1)

# Tinklelis
plt.grid()

# Ašių skalės
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.show()
```
